```python
# Age Demographic
# Establish bins for age
# Categorize the existing players using the age bins
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold results
```

```python
# Create bins
bins = [0,9,14,19,24,29,34,39,99]

# Age Group
age_group = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
```

```python
purchase_age_df = player_df.groupby(['Age Group'])
player_count = purchase_age_df['SN'].count()

#purchase_age_df['SN'].count()
#total_player_count = []
#player_pct = []
# Create data frame of count
plyr_demogrp = pd.DataFrame(player_count)
plyr_demogrp

perct_plyr = []

for row in plyr_demogrp['SN']:
    perct_plyr.append(round((row/total_player *100),2))

plyr_demogrp['Percentage of Players'] = perct_plyr

plyr_demogrp = plyr_demogrp.rename(columns={'SN':'Total Count'})

plyr_demogrp
```

```python
#Most Popular Items
# Identify the 5 most popular items by purchase count, then list (in a table):
# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value

```

```python
# Create a new Data Frame with Item ID, Item Name and Price
popitem_df = purchase_df.loc[:,['Item ID','Item Name','Price']]
popitem_df.head()
```

```python
#Group Data Frame by Item ID and Item Name
most_item = popitem_df.groupby(['Item ID','Item Name'])
most_item.head()

#most_pop_item = most_item

most_pop_item = pd.DataFrame(most_item['Price'].count())
most_pop_item['Item Price'] = pd.DataFrame(most_item['Price'].mean())
most_pop_item['Total Purchase Value'] = pd.DataFrame(most_item['Price'].sum())

most_pop_item = most_pop_item.rename(columns={'Price':'Purchase Count'})

most_pop_item1 = most_pop_item

most_pop_item1 = most_pop_item.sort_values('Purchase Count',ascending=False)
most_pop_item1['Item Price'] = most_pop_item1['Item Price'].astype(float).map('${:,.2f}'.format)
most_pop_item1['Total Purchase Value'] = most_pop_item1['Total Purchase Value'].astype(float).map('${:,.2f}'.format)

most_pop_item1.head(5)
```

```python
# Most Profitable Item
# Sort the Above DataFrame by Total Purchase Value in Descending Order.
most_pop_item2 = most_pop_item.sort_values('Total Purchase Value',ascending=False)

most_pop_item2['Item Price'] = most_pop_item2['Item Price'].astype(float).map('${:,.2f}'.format)
most_pop_item2['Total Purchase Value'] = most_pop_item2['Total Purchase Value'].astype(float).map('${:,.2f}'.format)

most_pop_item2.head(5)
```
